```python
import io

# Capture df.info() output
buffer = io.StringIO()
train_churn_df.info(buf=buffer)
info_str = buffer.getvalue()

# Process the output to DataFrame
lines = info_str.splitlines()[3:-2]  # Remove the header and footer lines
info_data = [line.split() for line in lines]  # Split each line by whitespace

info_df = pd.DataFrame(info_data[2:])
```

```python
info_df.columns = ["Index", "column", "Non-Null Count", "non-null", "dtype"]
```

```python
info_df.drop(["Index", "non-null", "Non-Null Count"], axis=1, inplace=True)
```

```python
category_info = pd.read_csv("category_mapping.csv")
```

```python
schema = pd.merge(
    category_info, 
    info_df, 
    left_on="Column", 
    right_on="column", 
    how="inner"
).drop("Column", axis=1)

schema = schema[['column', 'dtype', 'Category', 'Meaning']]
```

```python
# Separate the first row
first_row = schema.iloc[0]
df = schema.iloc[1:]

# Sort the remaining DataFrame
df = df.sort_values(by='dtype')

# Re-insert the first row at the top
df = pd.concat([pd.DataFrame(first_row).T, df], ignore_index=True)
df.columns = ['Column', 'Data Type', 'Category', 'Meaning']
df
```
